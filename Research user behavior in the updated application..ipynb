Первичная обработка данных и исследование

```python
import pandas as pd
events = pd.read_csv("7_4_Events (1).csv")
```

```python
purchase = pd.read_csv("7_4_Purchase (1).csv")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#проверяем данные 
events.info()
purchase.info()
```

```python
#фильтруем данные таблицы events_df по пользователями, которые зарегистрировались в период с 2018-01-01 по 2018-01-01
mask1 = events['start_time'] >= '2018-01-01'
mask2 = events['start_time'] < '2019-01-01'
mask3 = events['event_type'] == 'registration'

mask_2018 = events[mask1 & mask2 & mask3]['user_id'].to_list()
events = events[events['user_id'].isin(mask_2018)]

events.info()
```

```python
#Проверяем 
print(events['start_time'].min())
print(events['start_time'].max())
```

```python
#Фильтруем данные таблицы  purchase_df по попользователям,которые зарегестрировались и совершали  в период с 2018-01-01 по 2018-01-01 и проверяем:
purchase = purchase[purchase['user_id'].isin(mask_2018)]

purchase.info()
```

В обеих таблицах присутствуют одинаковые столбцы (id), но они имеют разноe значение.  
В датафрейме event_df столбце id указывает на id события, в purchase_df столбец id указывает на id оплаты.  

Переименовываем столбцы при помощи функции .rename()

```python
events = events.rename(columns={'id': 'event_id'})

purchase = purchase.rename(columns={'id': 'purchase_id'})
purchase = purchase.rename(columns={'event_datetime': 'start_time'})
purchase['event_type'] = 'purchase'
```

```python
# Обьединяем таблицы 

total_events = pd.concat([events, purchase], sort=False)

total_events.head()
```

```python
# Переводим столбец event_date в правильный формат:
total_events.start_time = pd.to_datetime(total_events.start_time)
total_events.info()
```

```python
total_events = total_events.reset_index(drop=True).sort_values('start_time')
total_events
```

```python
#Разделим пользователей на группы:
# Все пользователи 
all_user_id = set(total_events['user_id'].unique())
#Пользователи, которые начали обучение
user_id_with_start_tutorial = set(total_events[total_events['event_type'] == 'tutorial_start']['user_id'].unique())
#Пользователи, которые закончили оучение
user_id_with_finish_tutorial = set(total_events[total_events['event_type'] == 'tutorial_start']['user_id'].unique())
#Пользователи, которые не начинали обучение
user_id_without_started_tutorial = all_user_id.difference(user_id_with_start_tutorial)
```

ЗАДАЧА 1. Определить, насколько обучение сокращает время прохождения этапов игры.  

ГИПОТЕЗА 1. По идее, должна быть разница в поведении групп, которые проходят и не проходят обучение. Но так ли это? Влияет ли обучение на скорость прохождения других этапов игры?

```python
# Создаем сводную таблицу events_df
events_df = total_events.pivot_table(
    values='start_time',
    index='user_id',
    columns='event_type',
    aggfunc='min'
)
# Убираем индексы в сводной таблицы
events_df = events_df.reset_index()
events_df
```

```python
# Создаем функцию get_tutorial_status , делим  по группам Закончил, Не начинал и Не закончил:
def get_tutorial_status(user_id):
    if user_id in user_id_with_finish_tutorial:
        return 'Закончил'
    elif user_id in user_id_without_started_tutorial:
        return 'Не начинал'
    else:
        return 'Не закончил'
    
#Создаем новый столбец 
events_df['group'] = events_df['user_id'].apply(get_tutorial_status)
events_df    
```

```python
#Создаем новые столбцы
#показывает разницу во времени между purchase и registreation
events_df['deltatime_reg_and_purchase'] = (events_df['purchase'] - events_df['registration']) / pd.Timedelta('1 hour')

#показывает разницу во времени между pack_choice и registreation
events_df['deltatime_reg_and_pack_choice'] = (events_df['pack_choice'] - events_df['registration']) /  pd.Timedelta('1 hour')

#показывает разницу во времени между level_choice и registreation
events_df['deltatime_reg_and_level_choice'] = (events_df['level_choice'] - events_df['registration']) / pd.Timedelta('1 hour')

display(events_df)
```

Группировка данных по среднему времени

```python
#между регистрацией и покупкой по группе пользователей
grouped_deltatime_reg_and_puchase = round(events_df.groupby('group', as_index=False)[['deltatime_reg_and_purchase']].mean(),2)

#между регистрацией и выбором пакета вопросов по группе пользователей
grouped_deltatime_reg_and_pack_choice = round(events_df.groupby('group', as_index=False)[['deltatime_reg_and_pack_choice']].mean(),2)

#между регистрацией и выбором уровня сложности по группе пользователей
grouped_deltatime_reg_and_level_choice = round(events_df.groupby('group', as_index=False)[['deltatime_reg_and_level_choice']].mean(),2)
```

Сторим столбчатые диаграммы по группам пользователей 

```python
# Диаграмма 1 - показывает средню разницу во времени между регистрации и покупкой

fig = plt.figure(figsize=(8, 4))
axes = fig.add_axes([0,0,1,1])
bar_group_1 = axes.bar(
    x= grouped_deltatime_reg_and_puchase['group'],
    height = grouped_deltatime_reg_and_puchase['deltatime_reg_and_purchase']
)
axes.set_title('Время между регистрацией и покупкой')
axes.set_ylabel('Среднее время (час)')
axes.set_xlabel('Группа пользователей')
axes.bar_label(bar_group_1)
```

```python
# Диаграмма 2 - показывает среднюю разницу во времени между регистрацией и выбором пакета вопросов
fig = plt.figure(figsize=(8, 4))
axes = fig.add_axes([0,0,1,1])
bar_group_2 = axes.bar(
    x =  grouped_deltatime_reg_and_pack_choice['group'],
    height = grouped_deltatime_reg_and_pack_choice['deltatime_reg_and_pack_choice'])
axes.set_title('Время между регистрацией и выбором пакета вопросов')
axes.set_ylabel('Среднее время (час)')
axes.set_xlabel('Группа пользователей')
axes.bar_label(bar_group_2)
```

```python
#Диаграмма 3 - показывает среднюю разницу во времени между регистрацией и выбором уровня сложности 

fig = plt.figure(figsize=(8, 4))
axes = fig.add_axes([0,0,1,1])
bar_group_3 = axes.bar(
    x =  grouped_deltatime_reg_and_level_choice['group'],
    height = grouped_deltatime_reg_and_level_choice['deltatime_reg_and_level_choice'])
axes.set_title('Время между регистрацией и выбором уровня сложности')
axes.set_ylabel('Среднее время (час)')
axes.set_xlabel('Группа пользователей')
axes.bar_label(bar_group_3)
```

ВЫВОДЫ:

Диаграмма 1: показывает, что пользователи, которые прошли обучение и закончили его в среднем  тратят меньше  времени между регистрацией и покупкой

Диаграмма 2:  показывает, что пользователи, которые не начинали обучение тратят меньше времени между регистрацией и выбором пакета вопросов 

Диаграмма 3: показывает, что пользователи, которые не проходили обучение в среднем тратят меньше времени от регистрации до выбора уровня сложности.   



ОТВЕТ НА ГИПОТЕЗУ 1 

Разница в поведении групп, которые проходят и не проходят обучение, видна. Пользователи, которые прошли обучение, тратят меньше времени между регистрацией и покупкой, но больше времени между выбором покета вопросов и уровнем сложности. ГИПОТЕЗА 3. Если пользователь сначала выбирает сложность, будет ли он потом проходить обучение?

```python
#Создаем датафрейм с пользователями, которые начали обучение (tutorial_start)
tutorial_satart_group = total_events[total_events['event_type'] == 'tutorial_start']

#Создаем датафрейм с пользователями, которые выбрали уровень сложности (level_choice)
level_choice_group = total_events[total_events['event_type']== 'level_choice']

#Переменуем столбец start_time
level_choice_group = level_choice_group.rename(columns = {'start_time': 'level_choice_time'})

#Обьединяем таблицы
delta_tutorial_start_level_choice = tutorial_satart_group[['user_id', 'start_time']].merge(
    level_choice_group[['user_id', 'level_choice_time']],
    on= 'user_id',
    how = 'inner'
)

##Создаем доп столбце delta_time которая показывает разницу во времени
delta_tutorial_start_level_choice['delta_time'] = (delta_tutorial_start_level_choice['start_time'] - delta_tutorial_start_level_choice['level_choice_time'])

delta_tutorial_start_level_choice.info() 
display(delta_tutorial_start_level_choice)
```

```python
#Рассчитываем пользователей, которые приступили к обучению после выбора уровня сложности

user_level_choice_before_tutorial = delta_tutorial_start_level_choice.query('delta_time > "1 second"')['user_id'].nunique()
print('Количество пользователей, приступивших в обучени. полсе выбора уровня сложности:', user_level_choice_before_tutorial)
```

```python
#Рассчитываем пользователей, которые приступили к  выбору уровня сложности после начала обучения

user_level_choice_after_tutorial = delta_tutorial_start_level_choice.query('delta_time < @pd.Timedelta(0)')['user_id'].nunique()
print('Количество пользователей, приступивших к выбору уровня сложности после начала обучения:', user_level_choice_after_tutorial)
```

```python
#Находим долю пользователей, которые начали обучение после выбора уровня сложности :

users = user_level_choice_before_tutorial / user_level_choice_after_tutorial
print('Доля пользователей, которые начали обучение после выбора уровня сложности: {:.2%}'.format(users))
```

ВЫВОД:  
В результате расчетов видим, что почти 17% пользователей начинают обучение после выбора уровня сложноcти, это говорит о том, что этим пользователям  не понятна работа с приложением.
